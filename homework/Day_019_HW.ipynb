```python
# 載入需要的套件
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns # 另一個繪圖-樣式套件

# 忽略警告訊息
%matplotlib inline
plt.style.use('ggplot')
import warnings
warnings.filterwarnings('ignore')

app_train=pd.read_csv(r"C:\Users\QQ\Desktop\training data\application_train.csv")
app_train.head()
print(app_train.shape)
```

```python
# 使用不同的 HOUSETYPE_MODE 類別繪製圖形, 並使用 subplot 排版
unique_house_type = app_train["HOUSETYPE_MODE"].unique()
print(len(unique_house_type))
#s=np.unique(b,return_index=True)中的return_index=True 表示返回新列表元素在舊列表中的位置，並以列表形式儲存在s中。
#unique函式去除其中重複的元素，並按元素由小到大返回一個新的無重複的列表
nrows = len(unique_house_type)#len()返回对象长度
ncols = 2
#" // "表示整数除法

plt.figure(figsize=(15,30))
for i in range(len(unique_house_type)):
    plt.subplot(nrows, ncols, i+1)
    #多合一顯示
    #plt.subplot(2,2,3)表示将整个图像窗口分为2行2列,当前位置为3
   
    app_train.loc[app_train['HOUSETYPE_MODE'] == unique_house_type[i], "AMT_CREDIT"].hist()#畫histogram
    
    plt.title(str(unique_house_type[i]))
plt.show()    

#NAN代表HOUSETYPE_MODE裡的空格
```
