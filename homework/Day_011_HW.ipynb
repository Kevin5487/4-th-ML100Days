```python
import pandas as pd
import numpy as np

df=pd.read_csv(r"C:\Users\QQ\Desktop\training data\application_train-d11.csv")
df.head()

# 1: 計算 AMT_ANNUITY 的 q0 - q100
q_all = [np.percentile(df[df['AMT_ANNUITY'].notnull()]['AMT_ANNUITY'], q=i) for i in range(0, 101)]

print(pd.DataFrame({'q': list(range(101)),
              'value': q_all}))

# 2.1 將 NA 以 q50 填補
print("Before replace NAs, numbers of row that AMT_ANNUITY is NAs: %i" % sum(df['AMT_ANNUITY'].isnull()))

q_50 =  np.percentile(df[df['AMT_ANNUITY'].notnull()]['AMT_ANNUITY'],50)#要用percentile!!
df.loc[df['AMT_ANNUITY'].isnull(),'AMT_ANNUITY'] = q_50 #注意寫法

print("After replace NAs, numbers of row that AMT_ANNUITY is NAs: %i" % sum(df['AMT_ANNUITY'].isnull()))

# 2.2 Normalize values to -1 to 1
print("== Original data range ==")
print(df['AMT_ANNUITY'].describe())

def normalize_value(x):#要加冒號!#注意定義函數時用法!
    return 2*((x-np.min(x))/(np.max(x)-np.min(x))-0.5)

df['AMT_ANNUITY_NORMALIZED'] = normalize_value(df['AMT_ANNUITY'])

print("== Normalized data range ==")
print(df['AMT_ANNUITY_NORMALIZED'].describe())


#3 NA填入眾數
print("Before replace NAs, numbers of row that AMT_GOODS_PRICE is NAs: %i" % sum(df['AMT_GOODS_PRICE'].isnull()))

# 列出重複最多的數值
print(df['AMT_GOODS_PRICE'].value_counts().head())#列出最多的前五個

mode_goods_price = list(df['AMT_GOODS_PRICE'].value_counts())
df.loc[df['AMT_GOODS_PRICE'].isnull(), 'AMT_GOODS_PRICE'] = mode_goods_price[0]#帶入最多的也就是眾數

print("After replace NAs, numbers of row that AMT_GOODS_PRICE is NAs: %i" % sum(df['AMT_GOODS_PRICE'].isnull()))
```
